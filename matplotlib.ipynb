# توضیحات بیشتر درمورد این کتابخونه

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# یک شکل کشیدن تابع نرمال

value = np.random.normal(2 , 0.5 ,1000)
plt.hist(value , 50)
plt.show()


#  شکل دیگر کشیدن تابع نرمال

value = np.arange(-10 , 10 , 0.1)
plt.plot(value , norm.pdf(value))
plt.show()
```
