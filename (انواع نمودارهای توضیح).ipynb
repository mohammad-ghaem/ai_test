```python
#نمودار توزیع نرمال

import numpy as np
import matplotlib.pyplot as pt

Disturbiotion_normal= np.random.normal(20000, 5000 , 40000)

pt.hist(Disturbiotion_normal,50)
pt.show()
```

# شکل دیگه نمودار تابع نرمال

(pdf) means probability density function

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm 

x=np.arange(-0 , 10 , 1)
plt.plot(x,norm.pdf(x))
```

# توضیح تابع axes 

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt 
value = np.arange(-1,5,0.5)


axes = plt.axes()
axes.set_xlim(1,5) #برای محدود کردن خط ایکس 
axes.set_ylim(0,5) #برای محدود کردن خط وای 


axes.set_xticks([1,2,3,4,5,6,7]) #برای تعیین میزان فاصله در ایکس
axes.set_yticks([1,2,3,4,5]) #برای تعیین میزان فاصله در وای

axes.grid()# برای شطرنجی کردن صفحه

plt.ylabel("hello fella") #برای نوشتن در خط وای
plt.xlabel("hello fella") #برای نوشتن در خط ایکس


plt.plot(value , norm.pdf(value,1) , "r.") #برای رنگ و شکل دهی نمودار
plt.show()

```
