```python
import matplotlib.pyplot as plt 

value = [20 , 30 , 40 ,24 , 10] # میزان سهم هر بخش را نشان میدهد
colors = [ 'r' ,'g' ,'b' , 'c' , 'y'] # رنک هر بخش را نشان میدهد
labels = ['iran' , 'japen' , 'chaina' , 'mashhad' , 'tehran'] # نام هر یک از آن سهم ها
explode= [0,0,0,0.2 ,0] #برامدگی هر قسمت را نشان میدهد

plt.pie(value ,labels =labels ,colors=colors  ,  explode =explode) # قسمت اصلی نمودار دایره ای
plt.show()
```

```python
import matplotlib.pyplot as plt 

values = [20 , 30 , 40 ,24 , 10] # میزان سهم هر بخش را نشان میدهد
colors = [ 'r' ,'g' ,'b' , 'c' , 'y'] # رنک هر بخش را نشان میدهد

plt.bar(range(0,5) ,values , color= colors)
plt.show()



```

# نمودار نقطه ای

# نمودار جعبه ای

```python
import numpy as np
import matplotlib.pyplot as plt 

value=np.random.randn(100) * 100 -1
high=np.random.randn(10)*20 - 3
low = np.random.randn(10)*-20 -3

data = np . concatenate((value ,high ,low)) #اینجا میگیم این داده هارو باهم ترکیب کن

plt.boxplot(data)
plt.show()
```
