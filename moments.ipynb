# moments
بحث مهم در آنالیز دیتا در آمار که شامل چهار بخش میانگین و واریانس و کشیدگی و چولگی میباشد

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sp


value = np.random.normal(1397,0.5,82000)
plt.hist(value , 100)
plt.show()
```

```python
np.mean(value)
```

```python
np.var(value)
```

# Skewness
Skewness = چولگی

```python
sp.skew(value)
```

# Kurtosis

Kurtosis = کشیدگی

```python
sp.kurtosis(value)
```
